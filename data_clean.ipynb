```python
import pandas as pd
```

```python
df= pd.read_csv('data/historical_air_quality_2021_vi.csv')
df
```

```python
df.columns
```

```python
# Xóa các cột Unnamed
df = df.loc[:,~df.columns.str.contains('Unnamed')]

# Xóa các cột không cần thiết
df = df.drop(['Trạng thái cảnh báo','Mức độ cảnh báo'],axis=1)

# Tách cột Vị trí thành Latitude và Longitude
df[['Latitude','Longitude']]= df['Vị trí'].str.split(',',expand=True)
df['Latitude']=pd.to_numeric(df['Latitude'],errors='coerce')
df['Longitude']=pd.to_numeric(df['Longitude'],errors='coerce')
df=df.drop('Vị trí',axis=1)

# Chuyển đổi các cột số sang định dạng số
df['Mã trạm'] = pd.to_numeric(df['Mã trạm'], errors='coerce')
df['Chỉ số AQI'] = pd.to_numeric(df['Chỉ số AQI'], errors='coerce')

# Xóa dòng không hợp lệ
df = df[
    (df['Mã trạm'] != 0) & 
    (df['Chỉ số AQI'] > 0) & 
    (df['Latitude'] != 0) & 
    (df['Longitude'].notna())
]

# Reset index sau khi xóa
df = df.reset_index(drop=True)

# Kiểm tra kết quả
print("Shape sau khi xóa:", df.shape)
df
```
